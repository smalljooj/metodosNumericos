```python
import numpy as np
import scipy.linalg as sla
import matplotlib.pyplot as plt
```

## Exercicio 2.1

```python
v = np.array([1, 2])
w = np.array([4, -6])

soma_v_w = v + w
subt_v_w = v - w

fig, axs = plt.subplots(1, 2, figsize = (6, 3))

axs[0].arrow(0, 0, v[0], v[1], head_width = 0.3, width = 0.1, color = [.3, .3, .6], length_includes_head = True)
axs[0].arrow(v[0], v[1], w[0], w[1], head_width = 0.3, width = 0.1, color = [.7, .7, .3], length_includes_head = True)
axs[0].arrow(0, 0, soma_v_w[0], soma_v_w[1], head_width = 0.2, width = 0.1, color = [.6, .8, .6], length_includes_head = True)

axs[0].legend(['v','w','v + w'])
axs[0].set_title('v + w')

axs[1].arrow(0, 0, v[0], v[1], head_width = 0.3, width = 0.1, color = [.3, .3, .6], length_includes_head = True)
axs[1].arrow(0, 0, w[0], w[1], head_width = 0.3, width = 0.1, color = [.7, .7, .3], length_includes_head = True)
axs[1].arrow(w[0], w[1], subt_v_w[0], subt_v_w[1], head_width = 0.3, width = 0.1, color = [.6, .8, .6], length_includes_head = True)

axs[1].legend(['v','w','v - w'])
axs[1].set_title('v - w')

plt.tight_layout()
plt.show()
```

## Exercicio 2.2

```python
def norm(x):
    total = 0
    for i in x:
        total += i*i
    return np.sqrt(total)
```

```python
print(norm(v), np.linalg.norm(v))
print(norm(w), np.linalg.norm(w))
print(norm(soma_v_w), np.linalg.norm(soma_v_w))
print(norm(subt_v_w), np.linalg.norm(subt_v_w))
```

## Exercicio 2.5

```python
def transpose(x):
    new_arr = []
    if x.shape == (len(x),):
        [new_arr.append([i]) for i in x ]
    elif x.shape == (1, len(x[0])):
        [new_arr.append([i]) for i in x[0] ]
    elif x.shape == (len(x), 1):
        new_arr.append([])
        [new_arr[0].append(i[0]) for i in x ]
    return np.array(new_arr)
```

```python
b = np.array([[0, 1]])
c = np.array([[0], [1]])
print(f'vetor comum transposto:\n {transpose(v)}')
print(f'vetor linha transposta:\n {transpose(b)}')
print(f'vetor coluna transposta:\n {transpose(c)}')
```

## Exercicio 2.8

```python

a = np.array([1, 2])
b = np.array([1.5, .5])
plt.figure(figsize = (3, 3))

p_beta = np.dot(a,b) / np.dot(a,a)

plt.arrow(0, 0, a[0], a[1], head_width = .1, width = .02, color = [.3, .4, .5], length_includes_head = True)
plt.arrow(0, 0, b[0], b[1], head_width = .1, width = .02, color = [.5, .4, .3], length_includes_head = True)

plt.plot([b[0] , p_beta * a[0]], [b[1], p_beta * a[1]], 'k--')
plt.plot(p_beta * a[0], p_beta * a[1], 'ko', markerfacecolor = 'w', markersize = 13)


plt.text((b[0] + p_beta * a[0])/2, (b[1] + p_beta * a[1]) / 2+.1, r'(b-$\beta$a)', fontweight = 'bold', fontsize = 10)

plt.show()
```

## Exercicio 2.9

```python
a = np.random.randn(2)
b = np.random.randn(2)

a_paralela = b * (np.dot(a, b) / np.dot(b, b))
a_perpendicular = a - a_paralela

plt.figure(figsize=(6,6))

plt.plot([0, a[0]], [0, a[1]], color = [.2, .4, .7], linewidth = 3, label = 'a')
plt.plot([0, b[0]], [0, b[1]], color = [.2, .6, .8], linewidth = 3, label = 'b')

plt.plot([0, a_paralela[0]], [0, a_paralela[1]], 'k--', linewidth = 3, label = 'a|')
plt.plot([0,a_perpendicular[0]],[0,a_perpendicular[1]], 'k:', linewidth = 3, label = 'a|_')

plt.axis('equal')
plt.legend()
plt.show()
```

## Exercicio 2.9

```python
a = np.random.randn(2)
b = np.random.randn(2)

a_paralela = b * (np.dot(a, b) / np.dot(a, a))
a_perpendicular = a - a_paralela

plt.figure(figsize=(6,6))

plt.plot([0, a[0]], [0, a[1]], color = 'k', linewidth = 3, label = 'a')
plt.plot([0, b[0]], [0, b[1]], color = [.7,.7,.7], linewidth = 3, label = 'b')

plt.plot([0, a_paralela[0]], [0, a_paralela[1]], 'k--', linewidth = 3, label = 'a|')
plt.plot([0,a_perpendicular[0]],[0,a_perpendicular[1]], 'k:', linewidth = 3, label = 'a|_')

plt.axis('equal')
plt.legend()
plt.show()
```

```python
bug_a_paralela = b * (np.dot(a, b) / np.dot(b, b))
bug_a_perpendicular = a - bug_a_paralela

plt.figure(figsize=(6,6))

plt.plot([0, a[0]], [0, a[1]], color = 'k', linewidth = 3, label = 'a')
plt.plot([0, b[0]], [0, b[1]], color = [.7,.7,.7], linewidth = 3, label = 'b')

plt.plot([0, bug_a_paralela[0]], [0, bug_a_paralela[1]], 'k--', linewidth = 3, label = 'a|')
plt.plot([0, bug_a_perpendicular[0]],[0, bug_a_perpendicular[1]], 'k:', linewidth = 3, label = 'a|_')

plt.axis('equal')
plt.legend()
plt.show()
```
