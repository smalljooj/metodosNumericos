# Ex5 - Gradiente Descendente - Ismael Marinho

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
def gradient_descent(fd, x, a, it = -1, e = 1e-12):
    err = 1
    x_prev = x
    i = 0
    
    while(err > e):
        
        if it != -1:
            if it == i:
                return (x, i)
        x_prev = x
        x = x - a * fd(x)
        err = np.abs(x - x_prev)/np.abs(x)
        i += 1
    return (x, i)
```

```python
f = lambda x: 1/50*x**4 + 1/50*x**2 + 1/5*x
fd = lambda x: 2/25*x**3 + 1/25*x + 1/5

gradient_descent(fd, 2.5, 1, 25)
```

```python
x = np.linspace(-5, 5, 30)  

plt.grid()
plt.plot(x, f(x), [gradient_descent(fd, 2.5, 1, 25)[0], gradient_descent(fd, 2.5, 1, 25)[0]], [0,0], 'ro')  
```

```python
g = lambda x: 2*x**3 - 11.7*x**2 + 17.7*x - 5
gd = lambda x: 6*x**2 - 23.4*x + 17.7

x = np.linspace(0, 4, 30)

plt.grid()
plt.plot(x, g(x), [gradient_descent(gd, 4, 0.1)[0]], [g(gradient_descent(gd, 4, 0.1)[0])], 'ro')  
```
